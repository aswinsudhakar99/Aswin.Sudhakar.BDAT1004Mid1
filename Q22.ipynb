+---------------+---------+
| Column Name   | Type    |
+---------------+---------+
| id            | int     |
| recordDate    | date    |
| temperature   | int     |
+---------------+---------+
id is the column with unique values for this table.
This table contains information about the temperature on a certain day.
 
 

Write a solution to find all dates' Id with higher temperatures compared to its previous dates (yesterday).

Return the result table in any order.

Please use the following input Weather table for your solution. To receive full marks you will need to create the database, create the table, insert the data below and execute the SQL query.

 

Input: 
Weather table:
+----+------------+-------------+
| id | recordDate | temperature |
+----+------------+-------------+
| 1  | 2015-01-01 | 10          |
| 2  | 2015-01-02 | 25          |
| 3  | 2015-01-03 | 20          |
| 4  | 2015-01-04 | 30          |
+----+------------+-------------+
Output: 
+----+
| id |
+----+
| 2  |
| 4  |
+----+
Explanation: 
In 2015-01-02, the temperature was higher than the previous day (10 -> 25).
In 2015-01-04, the temperature was higher than the previous day (20 -> 30).

```python
import sqlite3
import pandas as pd

weather_db_conn = sqlite3.connect(':memory:')
weather_db_cursor = weather_db_conn.cursor()

create_weather_data_query = '''
CREATE TABLE Weather (
    weather_id INT PRIMARY KEY,
    weather_date DATE,
    weather_temperature INT
);
'''

weather_db_cursor.execute(create_weather_data_query)

insert_weather_data_query = '''
INSERT INTO Weather (weather_id, weather_date, weather_temperature) VALUES
(1, '2015-01-01', 10),
(2, '2015-01-02', 25),
(3, '2015-01-03', 20),
(4, '2015-01-04', 30);
'''

weather_db_cursor.execute(insert_weather_data_query)
weather_db_conn.commit()

weather_select_query = '''
SELECT W1.weather_id
FROM Weather W1
JOIN Weather W2
ON W1.weather_date = date(W2.weather_date, '+1 day')
WHERE W1.weather_temperature > W2.weather_temperature;
'''

weather_sql_result = weather_db_cursor.execute(weather_select_query).fetchall()

weather_df_result = pd.DataFrame(weather_sql_result, columns=['weather_id'])

print("+----+")
print("| id |")
print("+----+")
for row in weather_df_result['weather_id']:
    print(f"| {row}  |")
print("+----+")

weather_db_conn.close()
```
